# dtale for EDA and Visualization

```python
# import libraries
import pandas as pd
import seaborn as sns
import dtale
```

```python
# load data
df = sns.load_dataset('tips')
# start dtale server
d = dtale.show(df)
```

```python
# open browser
d.open_browser()
```

```python
# close dtale server
d.kill()
```
